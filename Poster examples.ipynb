```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.collections import PatchCollection
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
```

```python
font = "sans-serif"
fig = figure(figsize=(8,8))
fig.canvas.set_window_title('Instantaneous Field of View Diagram')
ax = plt.axes([0,0,1,1])

# create 3x3 grid to plot the artists
pos = np.mgrid[0.2:0.8:3j, 0.2:0.8:3j].reshape(2, -1)

patches = []
title('Instantaneous Field of View', fontsize = '20')
# add a circle
Veritas = mpatches.Circle(pos[:,3], 0.0114,ec="none")
patches.append(Veritas)
plt.text(pos[1,0]+0.3, pos[1,0]-0.04, "Veritas", ha="center",
        family=font, size=14)
plt.text(pos[1,0]+0.3, pos[1,0]-0.07, "~0.1% of the sky", ha="center",
        family=font, size=14)
HAWC = mpatches.Circle(pos[:,6], 0.2, ec="none")
patches.append(HAWC)
plt.text(pos[1,3]+0.6, pos[1,3] -0.02, "HAWC", ha="center",
        family=font, size=14)
plt.text(pos[1,3]+0.6, pos[1,3]-0.05, "~16% of the sky", ha="center",
        family=font, size=14)
HESS = mpatches.Circle(pos[:,8], 0.0239,ec="none")
patches.append(HESS)
plt.text(pos[1,0]+0.6, pos[1,0]+0.55, "HESS", ha="center",
        family=font, size=14)
plt.text(pos[1,0]+0.6, pos[1,3]+0.52, "~0.2% of the sky", ha="center",
        family=font, size=14)
HESS2 = mpatches.Circle(pos[:,7], 0.0114,ec="none")
patches.append(HESS2)
plt.text(pos[1,0]+0.6, pos[1,0]+0.26, "HESS2", ha="center",
        family=font, size=14)
plt.text(pos[1,0]+0.6, pos[1,0]+0.23, "~0.1% of the sky", ha="center",
        family=font, size=14)
CTA = mpatches.Circle(pos[:,0], 0.06,ec="none")
patches.append(CTA)
plt.text(pos[1,0], pos[1,0]-0.1, "CTA", ha="center",
        family=font, size=14)
plt.text(pos[1,0], pos[1,0]-0.125, "~0.5% of the sky", ha="center",
        family=font, size=14)

LAT = mpatches.Circle(pos[:,5]-0.14, 0.314,ec="none")
patches.append(LAT)
plt.text(pos[1,0]+0.165, pos[1,0]+0.45, "Fermi LAT", ha="center",
        family=font, size=14)
plt.text(pos[1,0]+0.165, pos[1,0]+0.42, "~20% of the sky", ha="center",
        family=font, size=14)

colors = 100*np.random.rand(len(patches))
collection = PatchCollection(patches, cmap=matplotlib.cm.jet, alpha=0.4)
collection.set_array(np.array(colors))
ax.add_collection(collection)

ax.set_xticks([])
ax.set_yticks([])

plt.show()
```
